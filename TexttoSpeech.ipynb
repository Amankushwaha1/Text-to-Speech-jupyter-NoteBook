```python
# Import the required module for text
# to speech conversion
from gtts import gTTS
from IPython.display import *
from langdetect import detect
#from textblob import TextBlob

# This module is imported so that we can
# play the converted audio
import os

# The text that you want to convert to audio
mytext = "Hi this is the Text-to-speech project done by Aman Kushwaha. This project uses google-text-to-speech library in python and can convert direct text, text in any form to mp3 file and autoplays the mp3 file"
#myfile = open("reading.txt","rt") 
#open file from system
#mytext2 = myfile.read()
```

```python
# Language in which you want to convert
language = 'en'

# Passing the text and language to the engine,
# here we have marked slow=False. Which tells
# the module that the converted audio should
# have a high speed
myobj = gTTS(text=mytext, lang=language, slow=False) #change text=mytext to mytext2 

# Saving the converted audio in a mp3 file named
# welcome
myobj.save("welcome.mp3")

# Playing the converted file
os.system("mpg321 welcome.mp3")

Audio("welcome.mp3", autoplay=True)
```
